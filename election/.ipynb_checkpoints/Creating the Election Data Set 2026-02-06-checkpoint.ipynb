This is a processing script to aggregate <a href="https://electionlab.mit.edu/data">MIT's Election Data</a> for United States presidental election at the state and county levels.  I use this data for teaching a Spatial Analysis course at Binghamton University.

```python
import pandas as pd
import numpy as np

import geopandas
```

# County Election Data

```python
mit_data = pd.read_csv('original_data/countypres_2000-2024.csv',dtype={'county_fips':str})
mit_data = mit_data.rename(columns={'county_fips':'FIPS'})
mit_data = mit_data[~mit_data['FIPS'].isnull()]
mit_data['FIPS'] = mit_data['FIPS'].str.zfill(5)
```

#### Data Repair: Reclassify Shannon County FIPS as Oglala Lakota County FIPS

```python
mit_data.loc[mit_data['FIPS']=='46113','FIPS'] = '46102'
```

#### Data Repair: Some counties in 2020 list separate tallies for different kinds of ballots

```python
# Example
mit_data.loc[mit_data.county_name=='MANASSAS PARK CITY']
```

```python
# Pull just 2020
df = mit_data[mit_data['year'] == 2020].copy()

# Add a field to count, when we sum this, it will tell us how many rows there were
df['count'] = 1

# Sum the votes by fips code and candidate
out = (
    df
    .groupby(['FIPS', 'candidate'], as_index=False)
    .agg({
        'candidatevotes': 'sum',
        'totalvotes': 'sum'
    })
)

out['year'] = 2020
```

```python
# Check to see if this looks right
out.loc[out.FIPS=='51685']
```

```python
mit_data = mit_data[mit_data['year'] != 2020]

mit_data = pd.concat(
    [mit_data, out],
    ignore_index=True
)

mit_data.tail()
```

```python
# Pull just 2024
df = mit_data[mit_data['year'] == 2024].copy()

# Add a field to count, when we sum this, it will tell us how many rows there were
df['count'] = 1

# Sum the votes by fips code and candidate
out = (
    df
    .groupby(['FIPS', 'candidate'], as_index=False)
    .agg({
        'candidatevotes': 'sum',
        'totalvotes': 'sum'
    })
)

out['year'] = 2024
```

```python
# Check to see if this looks right
out.loc[out.FIPS=='51685']
```

```python
mit_data = mit_data[mit_data['year'] != 2024]

mit_data = pd.concat(
    [mit_data, out],
    ignore_index=True
)

mit_data.tail()
```

#### Continue with data processing

```python
presidential_candidates = {2000:{'gop':'GEORGE W. BUSH','dem':'AL GORE'},
                           2004:{'gop':'GEORGE W. BUSH','dem':'JOHN KERRY'},
                           2008:{'gop':'JOHN MCCAIN','dem':'BARACK OBAMA'},
                           2012:{'gop':'MITT ROMNEY','dem':'BARACK OBAMA'},
                           2016:{'gop':'DONALD TRUMP','dem':'HILLARY CLINTON'},
                           2020:{'gop':'DONALD J TRUMP','dem':'JOSEPH R BIDEN JR'},
                           2024:{'gop':'DONALD J TRUMP','dem':'KAMALA D HARRIS'}
                           }
```

```python
output_df = pd.DataFrame()
output_df['FIPS'] = mit_data['FIPS'].unique()

years = np.sort(list(presidential_candidates.keys()))

for year in years:
    # Pull this year as a dataframe, pull this year's candidates, and 
    # convert year to a string, since it will now be used to name fields
    df=mit_data[mit_data['year']==year]
    candidates = presidential_candidates[year]
    year = str(year)

    # Get candidate info for this year, rename
    gop = df.candidate == candidates['gop']
    gop = df.loc[gop,['FIPS','candidatevotes']]
    gop = gop.rename(columns={'candidatevotes':'gop' + '_' + year + '_votes'})
    dem = df.candidate == candidates['dem']
    dem = df.loc[dem,['FIPS','candidatevotes','totalvotes']]
    dem = dem.rename(columns={'candidatevotes':'dem' + '_' + year + '_votes'})
    dem = dem.rename(columns={'totalvotes':'totalvotes' + '_' + year}) 
    
    # Write this information to the output dataframe and calculate some fields
    output_df = output_df.merge(gop,on='FIPS',how='left')
    output_df = output_df.merge(dem,on='FIPS',how='left')
    output_df['gop_' + year + '_prc'] = np.round(100 * output_df['gop_' + year + '_votes'] / output_df['totalvotes_' + year],decimals=2)
    output_df['dem_' + year + '_prc'] = np.round(100 * output_df['dem_' + year + '_votes'] / output_df['totalvotes_' + year],decimals=2)
    output_df['gop_minus_dem_prc_' + year] = output_df['gop_' + year + '_prc'] - output_df['dem_' + year + '_prc']
    
output_df.head()
```

```python
# Fix for [Kalawao County, Hawaii](https://en.wikipedia.org/wiki/Kalawao_County,_Hawaii), which doesn't have data.
# GEOID = 15005
# https://en.wikipedia.org/wiki/Kalawao_County,_Hawaii#Politics
# gop_minus_dem_prc_2020 = 91.66 with 24 total votes (of which 1 was republican)

d = {'FIPS':['15005'],'gop_2020_votes':[1],'dem_2020_votes':[23],'totalvotes_2020':[24],
                       'gop_2016_votes':[1],'dem_2016_votes':[14],'totalvotes_2016':[20],
                       'gop_2012_votes':[2],'dem_2012_votes':[25],'totalvotes_2012':[27],
                       'gop_2008_votes':[6],'dem_2008_votes':[24],'totalvotes_2008':[31],
                       'gop_2004_votes':[14],'dem_2004_votes':[26],'totalvotes_2004':[40],
                       'gop_2000_votes':[11],'dem_2000_votes':[30],'totalvotes_2000':[45],}
 
new_row = pd.DataFrame(d)

for year in [2000,2004,2008,2012,2016,2020]:
    year = str(year)
    new_row['gop_' + year + '_prc'] = np.round(100 * new_row['gop_' + year + '_votes'] / new_row['totalvotes_' + year],decimals=2)
    new_row['dem_' + year + '_prc'] = np.round(100 * new_row['dem_' + year + '_votes'] / new_row['totalvotes_' + year],decimals=2)
    new_row['gop_minus_dem_prc_' + year] = new_row['gop_' + year + '_prc'] - new_row['dem_' + year + '_prc']
    
output_df = pd.concat([output_df, new_row], ignore_index=True)
```

```python
output_df.to_csv('county_election_data_2000-2024.csv',index=False,float_format='%.2f')
```

# State Election Data

```python
mit_data = pd.read_csv('county_election_data_2000-2024.csv',dtype={'FIPS':str})
mit_data['FIPS'] = mit_data['FIPS'].str.zfill(5).str[:2]
```

```python
mit_data.head()
```

```python
# All totals are right at this point, but percentages are not
group = mit_data.groupby(by=['FIPS'])
mit_data = group.sum().reset_index()
```

```python
years = np.arange(2000,2025,4)
parties = ['gop','dem']
for year in years:
    for party in parties:
        mit_data[party + '_' + str(year) + '_prc'] = 100 * mit_data[party + '_' + str(year) + '_votes'] / mit_data['totalvotes_' + str(year)]
        if party=='dem':
            mit_data['gop_minus_dem_prc_' + str(year)] = mit_data['gop_' + str(year) + '_prc'] - mit_data['dem_' + str(year) + '_prc']
```

```python
mit_data.head()
```

```python
mit_data.to_csv('state_election_data_1976-2024.csv',index=False,float_format='%.2f')   
```

# Creating a County GeoPackage

```python
fn = "zip://original_data/cb_2020_us_county_20m.zip"
county_df = geopandas.read_file(fn,dtype={'GEOID':str})
county_df = county_df.drop(columns=['STATEFP','COUNTYFP','COUNTYNS','AFFGEOID','ALAND','AWATER','LSAD','NAMELSAD'])
county_df.head()
```

```python
fn = "zip://original_data/cb_2020_02_sldl_500k.zip"
ak_df = geopandas.read_file(fn,dtype={'GEOID':str})
ak_df = ak_df.drop(columns=['STATEFP','SLDLST','AFFGEOID','NAME','ALAND','AWATER','LSAD','LSY'])
ak_df = ak_df.rename(columns={'NAMELSAD': 'NAME'})
ak_df.head()
```

```python
county_df = county_df[county_df['STUSPS']!='AK']
```

```python
county_df = pd.concat([county_df,ak_df])
county_df.head()
```

```python
election_df = pd.read_csv('county_election_data_2000-2024.csv',dtype={'FIPS':str})
election_df.head()
```

```python
county_df = county_df.merge(election_df,how='left',left_on='GEOID',right_on='FIPS')
```

```python
county_df.to_file("election.gpkg", layer='county', driver="GPKG")
```

# Creating a CONUS GeoPackage

```python
fn = "zip://original_data/cb_2020_us_state_20m.zip"
conus_df = geopandas.read_file(fn,dtype={'GEOID':str})
conus_df = conus_df.drop(columns=['STATEFP','ALAND','AWATER','LSAD','STATENS','AFFGEOID','GEOID'])
conus_df = conus_df[~conus_df['STUSPS'].isin(['PR','AK','HI'])]
conus_df.head()
```

```python
conus_df.plot()
```

```python
election_df = pd.read_csv('state_election_data_1976-2024.csv',dtype={'FIPS':str})
election_df.head()
```

```python
# FIPS → postal code mapping
fips_to_postal = {
    "01": "AL", "02": "AK", "04": "AZ", "05": "AR", "06": "CA",
    "08": "CO", "09": "CT", "10": "DE", "11": "DC", "12": "FL",
    "13": "GA", "15": "HI", "16": "ID", "17": "IL", "18": "IN",
    "19": "IA", "20": "KS", "21": "KY", "22": "LA", "23": "ME",
    "24": "MD", "25": "MA", "26": "MI", "27": "MN", "28": "MS",
    "29": "MO", "30": "MT", "31": "NE", "32": "NV", "33": "NH",
    "34": "NJ", "35": "NM", "36": "NY", "37": "NC", "38": "ND",
    "39": "OH", "40": "OK", "41": "OR", "42": "PA", "44": "RI",
    "45": "SC", "46": "SD", "47": "TN", "48": "TX", "49": "UT",
    "50": "VT", "51": "VA", "53": "WA", "54": "WV", "55": "WI",
    "56": "WY"
}

# Ensure FIPS is clean, then map
election_df['STUSPS'] = (
    election_df['FIPS']
    .astype(str)
    .str.zfill(2)
    .map(fips_to_postal)
)
```

```python
election_df
```

```python
conus_df = conus_df.merge(election_df,how='left',left_on='STUSPS',right_on='STUSPS')
```

```python
conus_df.to_file("election_conus.gpkg", layer='state', driver="GPKG")
```
